```python
wid = pd.read_csv('C:/Users/Максим/Desktop/rahd/rshd3kt/World Important Dates.csv', delimiter=',')
```

```python
wid
```

```python
wid['Date'].unique()
```

```python
wid['Date'] = wid['Date'].replace('Unknown', pd.NA)
wid['Date'] = pd.to_numeric(wid['Date'], errors='coerce').fillna(0).astype(int)
wid['Date'] = wid['Date'].replace(1956, 0)
wid['Date'] = wid['Date'].replace(1998, 0)
wid['Date'] = wid['Date'].replace(1972, 0)
wid['Date'] = wid['Date'].replace(2004, 0)
wid['Date'] = wid['Date'].replace(1904, 0)
wid['Date'] = wid['Date'].replace(1917, 0)
wid['Date'] = wid['Date'].replace(1936, 0)
wid['Date'] = wid['Date'].replace(264, 0)
wid['Date'] = wid['Date'].replace(509, 0)
```

```python
wid['Date'].unique()
```

```python
wid.insert(wid.columns.get_loc('Year') + 1, 'Era', wid['Year'].apply(lambda x: 'BC' if 'BC' in x else 'AD'))
wid['Year'] = wid['Year'].str.replace('BC', '')
wid['Year'] = wid['Year'].astype(int)
```

```python
wid['Month'].unique()
```

```python
wid
```

```python
wid.to_parquet('World Important Dates.parquet.gzip',compression='gzip')
```
